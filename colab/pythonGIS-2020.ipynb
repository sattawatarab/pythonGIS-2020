4. ลิสต์ (list) 

คำสั้นๆ อธิบายก็สั้นๆ ก็คือ data structure ที่เก็บข้อมูลเป็นลำดับ (indexing) โดยลำดับแรกจะเริ่มที่ลำดับที่ศูนย์ (index = 0) 

และเพิ่มขึ้นเรื่อยๆ จนถึงลำดับสุดท้าย (index = :) การสร้าง list ทำได้ตาม 

ตัวอย่าง ข้างล่าง


````
list1 = []
print(type(list1))
#<class 'list'>
list_name = ['Gibbs', 'Donald', 'Jimmy', 'Abby']
list_age = [32, 32, 29, 31]
list_salary = [30000.00, 41309.50, 50230.01, 22949.80]
print(list_name)
#['Gibbs', 'Donald', 'Jimmy', 'Abby']
print(list_age)
#[32, 32, 29, 31]
print(list_salary)
#[30010.95, 41309.5, 50230.01, 22949.8]
print(len(list_name))
#4
````

จากที่บอกไปในข้างต้นว่า “List คือ data structure ที่เก็บข้อมูลเป็นลำดับ” ดังนั้นการสร้างหรือดึงข้อมูลออก จะใช้ลำดับ (indexing) เป็นตัวอ้างอิงทุกครั้ง

````
print(list_name[0])
#Gibbs
print(list_age[1])
#32
print(list_salary[3])
#22949.8
print(list_name[-1])
#Abby
print(list_name[-2])
#Jimmy
````

การดึงข้อมูลออกมาเป็นกลุ่ม เรียกอีกอย่างว่า “slicing” หรือการหั่นข้อมูลออกมา ทำได้โดยการใส่ [x,y] ต่อจาก list 

โดย x จะเป็น index เริ่มต้นที่ต้องการดึง และ y จะเป็น index ที่ต้องการหั่นออก ลองสังเกตุตัวอย่างด้านล่างดู

````
print(list_name[0:2])
#['Gibbs', 'Donald']
print(list_name[-3:-1])
#['Donald', 'Jimmy']
print(list_name[1:])
#['Donald', 'Jimmy', 'Abby']
print(list_name[:-2])
#['Gibbs', 'Donald']
````

5. dictionary

ต่อมาว่าด้วยเรื่องของ data structure อีกประเภทหนึ่งที่ใช้กันอย่างแพร่หลาย ชื่อว่า “dictionary”

โดย dictionary จะเก็บข้อมูลโดยไม่มีการอ้างอิงลำดับเหมือนกับ list แต่จะใช้ key เป็นตัวอ้างอิงแทน

````
dict1={'name':'Gibbs','age':32, 'salary':30010.95}
print(type(dict1))
#<class 'dict'>
print(dict1)
#{'name': 'Gibbs', 'age': 32, 'salary': 30010.95}
````

เรามาลองดึงค่าจาก dict1 ดูนะครับว่าเป็นยังไง จากเดิมถ้าเราดึงค่าจาก list เราจะอ้างอิงจากตำแหน่งของข้อมูลใช้มั๊ยครับ 

แต่ว่าของ dictionary นั้นเราทำได้ง่ายๆ โดยอ้างอิงจาก key โดยใช้ [key] ตามหลัง dictionary ลองดูตามตัวอย่างนะครับ

````
dict1={'name':'Gibbs','age':32, 'salary':30010.95}
print(dict1['name'])
#'Gibbs'
````

ยังไม่จบ นอกจาก list และ dictionary จะเก็บข้อมูลเดี่ยวๆ ได้แล้ว 

ก็ยังสามารถเก็บ list หรือ dictionary รวมกันได้ ตัวอย่างเช่น


````
dict2 = {'data':{'name':['Gibbs', 'Donald', 'Jimmy', 'Abby'], 'age':[32, 32, 29, 31], 'height':[180, 188, 167,170]}}
print(dict2['data']['age'])
#[32, 32, 29, 31]
print(dict2['data']['age'][1])
#32
````

2. intersection
- การหาส่วนซ้อนทับของ geometry จากสองชั้นข้อมูล

````
#selection data
povince_ns = polygon[(polygon['ProvinceTh'] == 'นครศรีธรรมราช')]
povince_ns.plot()
````

````
inter = povince_ns['geometry'].intersection(line['geometry'].unary_union)
inter.to_file('outdata.shp')
inter_red = gpd.read_file('outdata.shp', encoding = 'utf-8')
inter_red.plot()
````

````
f, ax = plt.subplots(figsize=(12, 6))
ax.set_title('Energy office near railway')

povince_ns.plot(ax=ax, facecolor='lightgray')
inter_red.plot(ax=ax)

res_bounds = inter.geometry.bounds

plt.xlim([res_bounds.minx.min()-0.5, res_bounds.maxx.max()+0.5])
plt.ylim([res_bounds.miny.min()-0.2, res_bounds.maxy.max()+0.8])
````

3. Buffer
- การสร้าง polygon แสดงขอบเขต จากการกำหนดระยะห่างออกจาก geometry object

````
#แปลงพิกัด จาก lat lon  ไป UTM
line_utm = line.to_crs({'init': 'epsg:32647'})
# สร้าง buffer ระยะห่างจากทางรถไฟ 20 km
line_20km = line_utm.buffer(20000)
line_20km.plot()
````

````
point_utm = point.to_crs({'init': 'epsg:32647'})
# นำ output จาก buffer มาใช้ทำ spatial query
result = point_utm[(point_utm.within(line_20km.geometry)) ]
result
````

````

f, ax = plt.subplots(figsize=(12, 6))
ax.set_title('Energy office near railway')

line_20km.plot(ax=ax, facecolor='lightgray')
line_utm.plot(ax=ax)

result.plot(ax=ax,marker='*', color='red', markersize=15)

res_bounds = result.geometry.bounds

plt.xlim([res_bounds.minx.min()-70000, res_bounds.maxx.max()+70000])
plt.ylim([res_bounds.miny.min()-70000, res_bounds.maxy.max()+70000])
````

2. Calculate NDVI

คำนวณหาค่า NDVI เพื่อดูค่าความสมบูรณ์ของพืชและพื้นที่ที่ไม่ใช่พืช โดยใช้ข้อมูล Landsat ถ

````
import rasterio
import rasterio.plot
import pyproj
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
````

````
#NDVI = (NIR - Red) / (NIR + Red)
b3 = rasterio.open('b-3.tif')
b4 = rasterio.open('b-4.tif')
red = b3.read()
nir = b4.read()
np.seterr(divide='ignore', invalid='ignore')
# Calculate ndvi

ndvi = (nir.astype(float)-red.astype(float))/(nir.astype(float)+red.astype(float))

ep.plot_bands(ndvi, 
              cmap='PiYG',
              scale=False,
              vmin=-1, vmax=1,
              title="NDVI, Color")
plt.show()
````

````
# Write the NDVI image
meta = b4.meta
meta.update(driver='GTiff')
meta.update(dtype=rasterio.float32)

with rasterio.open('NDVI.tif', 'w', **meta) as dst:
    dst.write(ndvi.astype(rasterio.float32))
````

````
#View Distribution of NDVI Values
ep.hist(ndvi,
        figsize=(12, 6),
        title=["NDVI: Distribution of pixels"])

plt.show()
````
